```python
import pandas as pd
import numpy as np

# Create Raw Healthcare Data (The Messy Part)
rows = 500
data = {
    'visit_id': range(100, 100 + rows),
    'dept': np.random.choice(['ER', 'Oncology', 'Cardio', 'ER_Dept', 'ONC', 'Cardiology'], rows), # Inconsistent naming
    'billing_amt': np.random.uniform(100, 5000, rows),
    'status': np.random.choice(['paid', 'PAID', 'pending', 'denied', 'DENIED', None], rows), # Messy strings + Nulls
    'wait_time': np.random.normal(60, 20, rows),
    'visit_date': pd.date_range(start='2023-01-01', periods=rows, freq='D')
}
df_client = pd.DataFrame(data)
df_client.to_csv('client_raw_data.csv', index=False)

# Create Supaboard Product Data (The internal analytics part)
users = 100
prod_data = {
    'user_id': range(1, users + 1),
    'sources_connected': np.random.randint(1, 10, users),
    'queries_asked': np.random.randint(0, 150, users),
    'is_retained': np.random.choice([0, 1], users, p=[0.4, 0.6])
}
df_prod = pd.DataFrame(prod_data)
df_prod.to_csv('supaboard_usage.csv', index=False)

print("Files 'client_raw_data.csv' and 'supaboard_usage.csv' have been created!")
```
